```python
#your task is to crate a circle constructor that creates a circle with a radius provided by an argument.the circles constructed must have two methods getArea() (pIr^2) and
#getPerimeter( 2PI*r) which give both respective areas and perimeter(Circumference)
class Circle:
  def __init__(self,radius):
    self.radius=radius
  def getArea(self):
    return 3.14*self.radius*self.radius
  def getPerimeter(self):
    return 2*3.14*self.radius
a=Circle(7)
print("Area of The circle=",a.getArea())
print("Perimeter of the circle=",a.getPerimeter())










```
